```python
#passengerId, name, Ticket sao valores diferentes, nao se usa no modelo de dados
#usando matplotlib
#histograma de tarifas
x = df['Fare']
# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=20, linewidth=0.5, edgecolor="white")

plt.show()
```

```python
x = df[df.Fare < 5].Fare
fig, ax = plt.subplots()
plt.title('Menor que 5 libras')

ax.hist(x, bins=40, linewidth=0.5, edgecolor="white")

plt.show()
```

```python
x = df[df.Fare > 100].Fare
fig, ax = plt.subplots()
plt.title('Maior que 100 libras')

ax.hist(x, bins=40, linewidth=0.5, edgecolor="white")

plt.show()
```
